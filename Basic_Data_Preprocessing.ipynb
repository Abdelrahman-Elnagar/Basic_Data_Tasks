```python
# a) Filter out rows with Juiciness higher than 0

filtered_data = dataset[dataset['Juiciness'] > 0]
filtered_data
```

```python
#b) Select the attribute that start with Q and print it and the count of its instances them.
##### we made an adjustment where we search for the column that start with q where the dataset doesnt have any string attributers to start with T
for column in dataset.columns:
    if column.startswith('Q'):
        quality_column = dataset[column]  # Selecting the column from the dataset
quality_counts = quality_column.value_counts()
print(quality_counts)

```

```python
#c) Find the total number of duplicated rows and drop them if there are any.
duplicate_rows = dataset[dataset.duplicated()]
print(f"Number of duplicate rows: {len(duplicate_rows)}")
dataset.drop_duplicates(inplace=True)
print(f"Dataset shape after dropping duplicates: {dataset.shape}")
```

```python
#d) Change type of numerical column from int to object.
##### we made an adjustment to select a float (not int)
dataset['Ripeness'] = dataset['Ripeness'].astype('object')
print(dataset['Ripeness'].info)
```

```python
#e) Group the entire data set by any two categorical attribute.
##### we made an adjustment to create anew attribute where it categorize the instances with acidity >0 to be positive and <0 to be negative then add this column to a copy of the origional dataset

import pandas as pd
df = dataset.copy()
# Create a new attribute based on the Ripeness value
df['Ripeness_Category'] = df['Ripeness'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

# Add the new attribute to the DataFrame
df['Ripeness_Category'] = df['Ripeness'].apply(lambda x: 'Positive' if x > 0 else 'Negative')

# Print the updated DataFrame
print(df)

grouped_data = df.groupby(['Quality', 'Ripeness_Category'])
grouped_data.head()
```

```python
#f) Are there any missing values in the data set?
missing_values = dataset.isnull().sum()
print(missing_values)

if (missing_values == 0).all():
    print("No missing values")
else:
    print("there is missing values")

```

```python
# g)If there are any NaN values in any column, substitute these missing values with the mean of

numerical = dataset.select_dtypes(include=numpy.number).columns
for column in numerical:
    dataset[column].fillna(dataset[column].mean(), inplace=True)

categorical = dataset.select_dtypes(include='object').columns
for column in categorical:
    dataset[column].fillna(dataset[column].mode()[0], inplace=True)

```

```python
# h) Discretize Size into 6 groups using equal interval method and count the values in
# each bin/Group. Make sure intervals are sorted when counting them

import pandas as pd
# Discretize 'Size' into 6 groups using equal interval method
bin_edges = numpy.linspace(dataset['Size'].min(), dataset['Size'].max(), 7)
bin_labels = ['Group {}'.format(i) for i in range(1, 7)]
dataset['Size_Group'] = pd.cut(dataset['Size'], bins=bin_edges, labels=bin_labels)

# Count the values in each bin/group
size_group_counts = dataset['Size_Group'].value_counts().sort_index()

# Print the counts
print(size_group_counts)
```

```python
# m) Standardize the numerical features in your dataset using the StandardScaler.

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

# Fit the scaler to the numerical columns
numerical_columns = dataset.select_dtypes(include=['int64', 'float64']).columns
dataset[numerical_columns] = scaler.fit_transform(dataset[numerical_columns])

# Print the standardized data
dataset
```

```python
# n) Apply Principal Component Analysis (PCA) to the standardized dataset, retaining two components. PCA is a dimensionality reduction technique that captures the most significant variance in the data. Visualize both the original and the reduced dimensions.

import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

new_dataset = dataset.copy()
new_dataset.drop('Size_Group', axis=1, inplace=True)
new_dataset['Quality'] = new_dataset['Quality'].apply(lambda x: 1 if x == 'good' else 0)

# Apply PCA to the standardized dataset
pca = PCA(n_components=2)
pca_data = pca.fit_transform(new_dataset)

# Visualize the original data
plt.scatter(new_dataset['Ripeness'], new_dataset['Juiciness'])
plt.xlabel('Ripeness')
plt.ylabel('Juiciness')
plt.title('Original Data')
plt.show()

# Visualize the reduced dimensions
plt.scatter(pca_data[:, 0], pca_data[:, 1])
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Reduced Dimensions')
plt.show()

# Get the principal component vectors
u1 = pca.components_[0]
u2 = pca.components_[1]

# Create a DataFrame with the reduced features
reduced_features = pca.transform(new_dataset)
reduced_df = pd.DataFrame(reduced_features, columns=['PC1', 'PC2'])

# Visualize the projection onto u1 and u2
plt.figure(figsize=(8, 6))
plt.scatter(reduced_df['PC1'], reduced_df['PC2'], alpha=0.5)
plt.quiver(0, 0, u1[0], u1[1], angles='xy', scale_units='xy', scale=0.2, color='r', label='u1')
plt.quiver(0, 0, u2[0], u2[1], angles='xy', scale_units='xy', scale=0.2, color='b', label='u2')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('Projection onto Principal Components')
plt.legend()
plt.grid(True)
plt.show()
```
